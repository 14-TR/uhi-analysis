```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your NDVI/LST timeseries CSV
df = pd.read_csv("NDVI_LST_Timeseries.csv")

# Calculate percent change
df['NDVI_PCT_Change'] = df['Mean_NDVI'].pct_change() * 100
df['LST_PCT_Change'] = df['Mean_LST'].pct_change() * 100

# Plot percent change over time
plt.figure(figsize=(10,6))
plt.plot(df['Year'], df['NDVI_PCT_Change'], label='NDVI % Change', marker='o')
plt.plot(df['Year'], df['LST_PCT_Change'], label='LST % Change', marker='o')
plt.axhline(0, color='gray', linestyle='--')
plt.title("Year-over-Year % Change: NDVI & LST")
plt.xlabel("Year")
plt.ylabel("Percent Change (%)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
