```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("NDVI_LST_Timeseries.csv")

# Plot NDVI and LST on dual y-axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# NDVI Plot
ax1.set_xlabel('Year')
ax1.set_ylabel('Mean NDVI', color='green')
ax1.plot(df['Year'], df['Mean_NDVI'], marker='o', color='green', label='NDVI')
ax1.tick_params(axis='y', labelcolor='green')

# LST Plot (Secondary Y-axis)
ax2 = ax1.twinx()
ax2.set_ylabel('Mean LST (K)', color='red')
ax2.plot(df['Year'], df['Mean_LST'], marker='s', color='red', label='LST')
ax2.tick_params(axis='y', labelcolor='red')

# Title and grid
plt.title('NDVI and LST Time Series (2013–2024)')
fig.tight_layout()
plt.grid(True)
plt.show()
```
